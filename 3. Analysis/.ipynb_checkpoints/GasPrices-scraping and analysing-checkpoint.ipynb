## Cleaning Data <a class="anchor" id="cleaning-data"></a>

### Issue 1 <a class="anchor" id="i1"></a>

Not available prices('×,××') need to be converted to NaN. <br>
Commas need to be replacet with dots. <br>
The Dtype of the column need to be converted to integer.

```python
gas_prices_copy['price'].replace('×,××', np.nan, inplace=True)
gas_prices_copy['price'] = gas_prices_copy['price'].str.replace(',','.')
gas_prices_copy['price'] = gas_prices_copy['price'].astype('float')       
```

```python
import math
def column_filter_plot(data, column_name,x, y, figsize=[16,16], rows=1, columns =3):
    rows = math.ceil(data[column_name].unique().shape[0]/ columns)
    fig = plt.figure(figsize=figsize)
    z= 1
    for s in data[column_name].unique():
        data_sub = data[data[column_name]==s]
        plt.subplot(rows, columns, z)
        bins_y = np.arange(data[y].min(), data[y].max()+.05, .05)
        bins_x = data_sub[x].value_counts().index.sort_values()
        plt.hist2d(data=data_sub, x=x, y=y, bins=[bins_x, bins_y])
        plt.colorbar()
        plt.xticks([0,1,2,3,4,5,6],['Mo','Tu','We', 'Th','Fr',"Sa",'So'])
        plt.ylabel('Gas Price (euro)')
        plt.title(s)
        
        z += 1
        
    plt.tight_layout()
```

```python
column_filter_plot(gp, "station_id", "weekday", "price", [16,16])
```

```python
#What is the std deviation per weekday on average
#group prices per week and then per weekday

#Create week number column in gp
gp['weeknumber'] = gp['price_time'].dt.isocalendar().week
```

```python
gp['weeknumber'].value_counts()
```

```python
std_week_num_day = gp.groupby(['weeknumber', 'weekday'])['price'].std()
reindex_std = std_week_num_day.reset_index()
indexes = ["{}".format(x) for x in std_week_num_day.index]


reindex_std['indexes'] = indexes
reindex_std.set_index('indexes', inplace=True)
reindex_std.drop(['weeknumber', 'weekday'], axis = 1, inplace =True)

fig = plt.figure(figsize=[16,6])
plt.plot(reindex_std, color=c)
plt.xticks(rotation=90)
plt.xlabel('Week,WeekDay')
plt.ylabel('avg. Price Deviation during day')
plt.title('Price Deviation');
```

```python
#Max and min per measered day
price_dif = gp.groupby(['weeknumber', 'weekday']).agg({'price':['mean', 'min', 'max']})
price_dif[('price','diff')] = price_dif[('price','max')]  - price_dif[('price', 'min')]
```

```python
price_dif
```

```python
#average max price different per day per station
price_dif_st = gp.groupby(['weeknumber', 'weekday','station_id']).agg({'price':['mean', 'min', 'max']})
price_dif_st[('price','diff')] = price_dif[('price','max')]  - price_dif[('price', 'min')]
price_dif_st.groupby(['weeknumber','weekday']).mean()
```

```python
fig = plt.figure(figsize=[16,4])
sns.lineplot(data=price_dif_st.groupby(['weeknumber','weekday']).mean(), x='weekday', y=("price","diff"), color=c )
plt.title("avg Price difference")
plt.xlabel("Day of week")
plt.ylabel('Price difference (euro cents)')
plt.xticks([0,1,2,3,4,5,6],['Mo','Tu','We', 'Th','Fr',"Sa",'So']);
```
