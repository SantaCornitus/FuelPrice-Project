```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
from datetime import datetime
```

```python
time = str(datetime.now().strftime("%Y-%m-%d(%H:%M)"))
print(time)
```

```python
#1. Get html website
url="https://ich-tanke.de/tankstellen/super-e10/ort/herzogenrath/"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
file_name = "herzogenrath{}.html".format(str(datetime.now().strftime("%Y-%m-%d,%H.%M")))
with open(file_name, mode='wb') as file:
    file.write(response.content)

    
```

```python
#open website with bs4
with open(file_name, encoding='utf-8') as file:
    soup = BeautifulSoup(file, "lxml")
```

```python
#first_entry = soup.find_all("div" , {id:"colm"})
content_page = soup.find(id="colm").find_all("li")

content_page[0].find("em")['title']
```

```python
#looping to all gasstation content and saving info to list
content_page = soup.find(id="colm").find_all("li")
now_time = datetime.now()

df_list = []
for t in content_page:
    station_id = t["id"]
    station_name = t.find_all("a")[0].contents[0]
    street_loc =t.find_all("div")[1].find_all('p')[0].contents[0][:-len(" · ")]
    city = t.find_all("div")[1].find_all('p')[0].find("a").contents[0]
    price = t.find("span", title="Preisangabe in Euro").contents[0]
    is24_7 = True if len(t.find_all("div")[1].find_all('p')[1].contents) != 1 else False
    
    df_list.append({"station_id": station_id, 
                    "station_name": station_name, 
                    "street":street_loc, 
                    "city": city, 
                    "price": price, 
                    "24/7":is24_7, 
                    "price_time": t.find("em")["title"][len("letzte Preisänderung: "):-4],
                    "download_time":now_time })
df3 = pd.DataFrame(df_list)
    
```
